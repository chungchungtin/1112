# python list容器 物件操作

```python
# list 可以包很多東西, 
somewhat_list = ['file_path', True, 10.3]
somewhat_list_list = [somewhat_list, {'姓名': '唉呦味', '年齡': 18}, (256, 256, 3)]
print(somewhat_list_list)
```

```python
# list 可以做 indexing
print(somewhat_list_list[1])
# list 也可以做 slicing
print(somewhat_list_list[1:3])
```

```python
# 還可以用 self.append() 做附加
somewhat_list_list.append("附加的東東")
print(somewhat_list_list)
# 還有簡便的寫法
somewhat_list_list += ["再加點東東"]
print(somewhat_list_list)
```

```python
# 還能對 element 做重複
somewhat_list_list += [123]*3 # type: ignore
print(somewhat_list_list)
```

# numpy 很會做陣列處理

```python
import numpy as np  # <-- 通常像這樣匯入
# numpy.ndarray 的 element 只能是一種資料型態, 可以使用 self.dtype 來查看資料屬性
str_arr = np.array(['哀有為', '吳柏毅', '熊貓', '這樣點'])
print("字串陣列:", str_arr, str_arr.dtype)
float_arr = np.array([1.23, 0.87, 3.1415926])
print("浮點陣列:", float_arr, float_arr.dtype) 
```

```python
# 可以改變數值型態
float_arr = float_arr.astype('float16')
print("16bit 浮點陣列:", float_arr, float_arr.dtype) 
```

```python
# 陣列可以做數值運算!!
float_arr = float_arr * 10
print("10x 結果:", float_arr, float_arr.dtype)
mask_arr = float_arr >= 10
print("可以做布林運算:", mask_arr)
```

```python
# 更厲害的是 可以多維度 n-dimention, 用 self.shape 查看形狀屬性
three_d_arr = np.ones(shape=(5,4,3), dtype='int8')
print("陣列的形狀 =", three_d_arr.shape)
print(three_d_arr)
```

```python
# 一樣可以做 indexing
three_d_arr[0, 1, 1] = 2
print(three_d_arr)
```

```python
# 一樣可以做 slicing
three_d_arr[:, :, 1] = 3
print(three_d_arr)
```

```python
# 用 slicing 取 sub-array
sub_arr = three_d_arr[3, :, :]
print(sub_arr)
```

```python
# 直接數值運算 要考慮數值型態
print(sub_arr.astype('int8') * 100)
print(sub_arr.astype('int16') * 100)
```

# 圖片的應用

```python
from PIL import Image  # pillow 圖檔套件
import requests
from io import BytesIO
import matplotlib.pyplot as plt  # 畫圖工具

response = requests.get(url='https://imgcdn.cna.com.tw/www/WebPhotos/1024/20210406/1779x1334_391943153473.jpg')
image_file = BytesIO(response.content)
pil_image = Image.open(image_file)
plt.imshow(pil_image) # type: ignore
plt.plot()
```

```python
# PIL.image 可以轉為 ndarray
image_arr = np.asarray(pil_image)
print('image_arr shape(h, w, c) =', image_arr.shape, image_arr.dtype)
```

```python
# slicing sub area
sub_arr = image_arr[300:600, 200:600, :]
print("他叫車車", "shape =", sub_arr.shape)
plt.imshow(sub_arr)
plt.plot()
```

```python
# array 也可以做 mask
b_mask = sub_arr[:,:,2] < 100
rgb_mask = np.repeat(b_mask[:, :, np.newaxis], 3, axis=2)
masked_img = sub_arr * rgb_mask + 100

plt.figure(figsize=(12,6))
plt.subplot(121)
plt.imshow(b_mask)
plt.title(f"mask {b_mask.shape}")
plt.subplot(122)
plt.imshow(masked_img)
plt.title(f"masked_img {sub_arr.shape}")
plt.plot()
```
