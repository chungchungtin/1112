```python
from PIL import Image  # pillow 
import requests
from io import BytesIO
import matplotlib.pyplot as plt  # 
import numpy as np

response = requests.get(url='https://th.bing.com/th/id/R.45e36d84ca5455ffa99d990b81c1f0ac?rik=34YMLAiZFj0Lqw&riu=http%3a%2f%2fwww.yalcincakir.com.tr%2fimages%2fcontent%2fnasil-yapilir%2fcekim-teknikleri%2fcekim-teknikleri-7%2f7-8.jpg&ehk=4EefkeTF8leGQa20LHEtpmhu2yHtZPsWv457bMLcdfw%3d&risl=&pid=ImgRaw&r=0')
image_file = BytesIO(response.content)
pil_image = Image.open(image_file)
plt.imshow(pil_image)
plt.plot()
```

```python
# PIL.image  ndarray
image_arr = np.asarray(pil_image)
print('image_arr shape(h, w, c) =', image_arr.shape, image_arr.dtype)
```

```python
sub_arr = image_arr[100:400, :700, :]
print("kid", "shape =", sub_arr.shape)
plt.imshow(sub_arr)
plt.plot()
```

```python
b_mask = sub_arr[:,:,1] < 164
rgb_mask = np.repeat(b_mask[:, :, np.newaxis], 3, axis=2)
masked_img = sub_arr * rgb_mask + 164

plt.figure(figsize=(12,6))
plt.subplot(121)
plt.imshow(b_mask)
plt.title(f"mask {b_mask.shape}")
plt.subplot(122)
plt.imshow(masked_img)
plt.title(f"masked_img {sub_arr.shape}")
plt.plot()
```
